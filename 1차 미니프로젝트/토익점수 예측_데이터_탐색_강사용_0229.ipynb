```python
#[문제] 'data' 데이터프레임의 'Birth_Year' 컬럼의 연도별 개수를 Bar 차트로 그리세요.
##[해설] 'BY'열의 값을 계산하려면 date[BY]의 value_counts()함수로 횟수를 세고, plot(kind=bar)으로 막대그래프 생성
##[해설] sort_index() 메서드를 사용, 생일년도를 오름차순으로 정렬(기본값은 ascending=True, 오름차순)
```

```python
# DataFrame value_counts()와 plot() 함수 활용
# 대상 컬럼 : 'Birth_Year'
# plot 함수의 인자 : kind='bar'

data[BY].value_counts().sort_index(ascending=True).plot(kind='bar', x='BY', y='Values')
```

```python
# 또 다른 방법으로 Bar 차트 생성
# plt.xticks(':X축의 눈금을 설정하는 함수, unique() 메서드는 컬럼의 고유한 값들을 반환, 즉 X축의 틱 위치가 
# 연도 데이터의 고유한 값들로 설정, 텍스트가 겹치지 않도록 rotation=90 설정

plt.bar(data[BY].value_counts().index, data[BY].value_counts().values)
plt.xticks(ticks=data[BY].unique(), rotation=90)
plt.title('Birth Year 차트')
plt.xlabel('Birth Year')
plt.ylabel('Count')
plt.show
```
