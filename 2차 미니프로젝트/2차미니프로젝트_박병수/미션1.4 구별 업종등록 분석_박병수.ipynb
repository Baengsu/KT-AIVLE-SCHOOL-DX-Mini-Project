# 3.데이터 분석하기
+ KeyPoint : 데이터의 형태를 살펴보고 다양한 분석기법을 통해 모델링에 적합하도록 정제요소를 선별할 수 있다.
  * 데이터들의 패턴 탐색
  * 변수들간의 관계 파악

```python
# 시각화 한글폰트 설정
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

plt.rc('font', family='Malgun Gothic')
sns.set(font="Malgun Gothic",#"NanumGothicCoding", 
        rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
        style='darkgrid')        
# 쉼표(,) 제거 후 정수로 변환
seoul_business1['택시운송업.1'] = seoul_business['한식 일반 음식점업.1'].str.replace(',', '').astype(int)


```

#### [실습문제4] 데이터 분포 알아보기
* 여러 변수들을 다양한 그래프로 그려보고 인사이트를 도출해보세요.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,8))

plt.plot(seoul_business['자치구'], seoul_business['택시운송업.1'], color='r', linestyle='-', marker='o', label='택시운송업')
plt.plot(seoul_business['자치구'], seoul_business['한식 일반 음식점업.1'], color='green', linestyle='-', marker='o', label='한식 일반 음식점업')


plt.xticks(rotation=45)


plt.legend()
plt.xlabel('자치구')
plt.ylabel('업종')
plt.title('구별 업종등록 분석')
plt.xticks(rotation=45)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.plot(seoul_business1['자치구'], seoul_business1['일반 교과 학원.1'], color='#F4A261', linestyle='-', marker='o', label='일반 교과 학원')
#plt.plot(seoul_business1['자치구'], seoul_business1['기타주점업.1'], color='#2A9D8F', linestyle='-', marker='s',label='기타 주점업')

plt.legend()
plt.xlabel('자치구')
plt.ylabel('이동인구')
plt.title('구별 업종등록 분석')
plt.xticks(rotation=45)

# 위 그래프와 설정 한꺼번에 보여주기
plt.show()
```
